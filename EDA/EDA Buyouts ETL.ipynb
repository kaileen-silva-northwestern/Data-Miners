```python
# EXTRACT

conn = pymssql.connect(server='mysqlserver-432.database.windows.net', user='azureuser', password='msds432!', database='mySampleDatabase')
query = """
select
    case_number,
    pre_buyout_disclosure_declaration_date,
    buyout_agreement_date, buyout_amount, 
    unknown_amount, 
    address,
    other_consideration
from dbo.SF_Buyout_Agreements_Raw ;""" 

buyouts_data = pd.read_sql_query(query, conn)
conn.close()
```

```python
buyouts_data
```

```python
# TRANSFORM
buyouts_data['pre_buyout_disclosure_declaration_date'] = buyouts_data['pre_buyout_disclosure_declaration_date'].apply(lambda s: s[:10] if s != "nan" else None)
```

```python
# TRANSFORM
buyouts_data['buyout_agreement_date'] = buyouts_data['buyout_agreement_date'].apply(lambda s: s[:10] if s != "nan" else None)
```

```python
# TRANFORM
buyouts_data['buyout_amount'] = buyouts_data['buyout_amount'].apply(lambda s: float(s) if s != "nan" else None)
```

```python
# TRANSFORM
buyouts_data['unknown_amount'] = buyouts_data['unknown_amount'].apply(lambda s: True if s != "nan" else False)
```

```python
# TRANSFORM
buyouts_data['other_consideration'] = buyouts_data['other_consideration'].apply(lambda s: True if s != "nan" else False)
```

```python
buyouts_data['building_id'] = None
```

```python
buyouts_data['location_id'] = None
```

```python
buyouts_data = buyouts_data[['case_number','location_id','pre_buyout_disclosure_declaration_date','buyout_agreement_date','address','buyout_amount','unknown_amount','other_consideration','building_id']]
```

```python
# LOAD
with pymssql.connect(server='mysqlserver-432.database.windows.net', user='azureuser', password='msds432!', database='mySampleDatabase') as conn, conn.cursor() as cursor:
    sql = """INSERT INTO dbo.Rel_BuyoutAgreements (
  case_number,
  pre_buyout_disclosure_declaration_date,
  buyout_agreement_date,
  buyout_amount,
  unknown_amount,
  address,
  other_consideration,
  Location_ID,
  Building_ID)
VALUES (%s, %s, %s, %d, %d, %s, %d, %d, %d)"""
    records = [tuple(row) for row in buyouts_data.to_numpy()]
    cursor.executemany(sql, records)
    conn.commit()    
```
